```python
import random
import json
from typing import List, Optional

# Skema untuk nilai sensor
class Value:
    def __init__(self, CO2: Optional[List[Optional[int]]], DO: Optional[List[Optional[int]]], 
                 PM2dot5: Optional[List[Optional[int]]], Temp: Optional[List[Optional[int]]], 
                 Humidity: Optional[List[Optional[int]]]):
        self.CO2 = CO2
        self.DO = DO
        self.PM2dot5 = PM2dot5
        self.Temp = Temp
        self.Humidity = Humidity

def generate_random_sensor_data():
    num_data_points = 1500  # 60 seconds * 5 minutes

    # Generate lists of random integer values for each sensor
    CO2 = [random.randint(300, 1000) for _ in range(num_data_points)]
    DO = [random.randint(0, 100) for _ in range(num_data_points)]
    PM2dot5 = [random.randint(0, 150) for _ in range(num_data_points)]
    Temp = [random.randint(20, 40) for _ in range(num_data_points)]  # Integer temperature
    Humidity = [random.randint(30, 70) for _ in range(num_data_points)]  # Integer humidity

    # Create an instance of the Value class with the generated data
    sensor_data = Value(
        CO2=CO2,
        DO=DO,
        PM2dot5=PM2dot5,
        Temp=Temp,
        Humidity=Humidity
    )
    
    # Convert to a dictionary and output as JSON
    sensor_data_dict = {
        "CO2": sensor_data.CO2,
        "DO": sensor_data.DO,
        "PM2dot5": sensor_data.PM2dot5,
        "Temp": sensor_data.Temp,
        "Humidity": sensor_data.Humidity
    }
    
    print(json.dumps(sensor_data_dict, indent=4))

if __name__ == "__main__":
    generate_random_sensor_data()
```
